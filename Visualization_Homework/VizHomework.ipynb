```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

##Step 2: Loading the Data (Connecting with the Data Source and Importing the Data into a DataFrame)

```python
# Extract and load the CSV dataset into a DataFrame
data = '/content/Top_Highest_Openings.csv'
df = pd.read_csv(data, low_memory = False)
```

##Step 3: Generating the Graphs

```python
# Graph 1: Scatter Plot - Opening vs. Total Gross Earnings

# Generate the plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Opening'], df['Total Gross'], color='red')
plt.title('Relationship Between Opening Weekend and Total Gross Earnings (in Hundreds of Millions)')
plt.xlabel('Opening Weekend Earnings ($)')
plt.ylabel('Total Gross Earnings ($)')
plt.grid(True)
plt.show()
```

```python
# Graph 2: Bar Chart - Top 10 Movies by Opening Weekend Earnings

# Sort the data and select the top 10 movies by opening weekend earnings
top_opening_movies = df.nlargest(10, 'Opening')

# Generate the plot
plt.figure(figsize=(12, 8))
plt.bar(top_opening_movies['Release'], top_opening_movies['Opening'], color='green')
plt.xlabel('Movies')
plt.ylabel('Opening Weekend Earnings ($)')
plt.title('Top 10 Movies by Opening Weekend Earnings (in Hundreds of Millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Graph 3: Line Plot - Average Opening Earnings Over Time

# Convert 'Date' into a datetime format and extract the year
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Year'] = df['Date'].dt.year

# Calculate the average opening earnings by year
average_opening_by_year = df.groupby('Year')['Opening'].mean()

# Generate the plot
plt.figure(figsize=(14, 7))
plt.plot(average_opening_by_year.index, average_opening_by_year.values, marker='o', linestyle='-', color='blue')
plt.title('Average Opening Weekend Earnings by Year (in Tens of Millions)')
plt.xlabel('Year')
plt.ylabel('Average Opening Weekend Earnings ($)')
plt.grid(True)
plt.show()
```
